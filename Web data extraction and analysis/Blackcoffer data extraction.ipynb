```python
from selenium.webdriver.common.by import By
import selenium
from selenium.webdriver.common.action_chains import ActionChains
import time
from selenium import webdriver
from selenium.webdriver.support.select import Select
from bs4 import BeautifulSoup
import os
import pandas as pd
import csv
```

```python
driver = webdriver.Chrome()
```

```python
df = pd.read_excel("input.xlsx")

url_ids = df['URL_ID']
urls = df['URL']
```

```python
def save_article_to_file(url_id, title, article_text):
    file_name = f"{url_id}.txt"
    with open(file_name, 'w', encoding='utf-8') as f:
        f.write(f"Title: {title}\n\n{article_text}")
```

```python
def extract_article_text(url):
    
    driver.get(url)
    time.sleep(6)
    while True:
        try: 
            article_text = driver.find_element(By.CLASS_NAME,'td-post-content.tagdiv-type')
            title = driver.title
            return title,article_text.text
        except:
            print("Error! data has not extraxted")
        break
```

```python
while True:
    for url_id, url in zip(url_ids, urls):
        print(url_id,url)
        try:
            title, article_text = extract_article_text(url)
            save_article_to_file(url_id, title, article_text)
            print(f"Article {url_id} saved successfully.")
        except Exception as e:
            print(f"Failed to extract {url_id}: {e}")
    break
```
